````python
# DO NOT MODIFY NOR ADD CODE TO THIS CELL
df = pd.read_csv('https://raw.githubusercontent.com/jnin/information-systems/refs/heads/main/data/superheroes_complete.csv')
````

````python
df= df.dropna(subset= ['Alignment'])  # After checking the data, we realize that "y" has missing values so the model would not work.
df.head() # Too see the values of the new columns.
````

````python
na_count = df.isna().sum()
total_entries = len(df)

# We create a DataFrame with counts and percentages
na_summary = pd.DataFrame({
    'NA Count': na_count,
    'Percentage': round((na_count / total_entries)*100,2)
})
na_summary.head(10)
````

````python
# Dropping the irrelevant columns and the ones with too many missing values(such as Race, Skin, Eye Color and Hair Color)
# We believe that these columns are not important since they can make the analysis less accurate.
# 'Name' and 'Publisher' are also irrelevant to the analysis of alignment.

X=df.drop(columns=['Name','Publisher', 'Alignment',
                   'Race', 'Skin color','Hair color','Eye color']) 
y=df['Alignment']
X.head()
````

````python
import numpy as np
X = X.replace(-99, np.nan) ## Observing the dataset, we can see that Height and Weight columns have -99 as values.
# We replace them with NAN values to then impute them.
````

````python
# Now whe import the functions we need for the pipeline
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
````

````python
# First we create the dataframes that will be used in the pipeline
X_ordinal = X[['Power Level', 'Intelligence Level']]
X_numerical = X[['Height', 'Weight']]
X_categorical = X.drop(columns=['Power Level', 'Intelligence Level', 'Height', 'Weight'])

# To impute the missing values for the numerical columns we will use the mean strategy.
numerical_imputer= SimpleImputer(strategy="mean")

# To impute the missing values for the categorical columns we will use the most frequent strategy and the one hot encoder.
cat_pipeline = Pipeline([
    ("most_frequent", SimpleImputer(strategy="most_frequent")),
    ("one_hot_encoder", OneHotEncoder(sparse_output=False))

])

# To impute the missing values for the ordinal columns we will use the most frequent strategy and the ordinal encoder.
categories = [['Weak','Below Average','Average','Above Average','Extremely Powerful'],
               ['Low Intelligence','Average Intelligence','Smart','Genius','Super-Genius']]
ord_pipeline = Pipeline([
    ("most_frequent", SimpleImputer(strategy="most_frequent")),
    ("ord_encoder", OrdinalEncoder(categories=categories))
])

# Now we gather all the pipelines in a ColumnTransformer.
columntransformer_pipeline = ColumnTransformer(
    transformers=[
        ("num_imputer", numerical_imputer, X_numerical.columns),
        ("cat_pipeline", cat_pipeline, X_categorical.columns),
        ("ord_pipeline", ord_pipeline, X_ordinal.columns)
],
remainder="passthrough"
)

columntransformer_pipeline
````

````python
X_encoded= columntransformer_pipeline.fit_transform(X)
X_encoded
````

````python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_encoded)
X_scaled.head(5)
````

````python
pipe = Pipeline(steps=[
    ('transformer', columntransformer_pipeline),
    ('scaler', scaler)
])
X_pipe = pipe.fit_transform(X)
X_pipe.head()
````

## Testing the Model

Now we run the logistic regression to see the score of the model

````python
unique_values_y = list(set(y))
unique_values_y # checking that y has no Nan values.
````

````python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

model_pipeline = Pipeline(steps = [
    ("transformer", columntransformer_pipeline),
    ("scaler", StandardScaler()),
    ("classifier", LogisticRegression(max_iter=1000))
]
)

model_pipeline.fit(X, y)
model_pipeline.score(X, y)
````
